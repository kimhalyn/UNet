```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# numpy가 저장된 디렉토리 설정
result_dir = './drive/MyDrive/training_unet/results/numpy'
```

```python
# 해당 디렉토리 데이터 추출
lst_data = os.listdir(result_dir)

# 데이터 분리
lst_label = [f for f in lst_data if f.startswith('label')]
lst_input = [f for f in lst_data if f.startswith('input')]
lst_output = [f for f in lst_data if f.startswith('output')]

lst_label.sort()
lst_input.sort()
lst_output.sort()
```

```python
# 데이터 불러오기
id = 0

label = np.load(os.path.join(result_dir, lst_label[id]))
input = np.load(os.path.join(result_dir, lst_input[id]))
output = np.load(os.path.join(result_dir, lst_output[id]))
```

```python
# 출력
plt.subplot(131)
plt.imshow(input, cmap='gray')
plt.title('input')

plt.subplot(132)
plt.imshow(label, cmap='gray')
plt.title('label')

plt.subplot(133)
plt.imshow(output, cmap='gray')
plt.title('output')

plt.show()
```
